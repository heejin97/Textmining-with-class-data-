```python
import pandas as pd
df = pd.read_csv('./2020-1_2.csv', encoding='utf-8')
df[0:130]
```

```python
posts = df.get('structure') #+ " " + df.get('efforts')+" " + df.get('eval')
posts[0:130]
```

```python
stop_words = "은 이 것 등 더 를 교수님 학생 학우 강의 없음 학우님 가 이론 수업 월요일 화요일 수요일 목요일 금요일"
stop_words=stop_words.split(' ')
stop_words[0:130]
```

```python
from eunjeon import Mecab
tagger = Mecab()
```

```python
nouns = []
for post in posts:
    for noun in tagger.nouns(post):
        if noun not in stop_words:
            nouns.append(noun)
nouns[0:130]
```

```python
words = []
for post in posts:
    words.extend(tagger.pos(post))
words[0:130]
```

```python
verbs=[]
for word in words:
    if word[1]=='VV':
        verbs.append(word[0]+'다')
verbs[0:130]
```

```python
adjective=[]
for word in words:
    if word[1]=='VA':
        adjective.append(word[0]+'다')
adjective[0:130]
```

```python
from collections import Counter
counted_nouns = Counter(nouns)
top_20 = counted_nouns.most_common(20)
top_20
```
